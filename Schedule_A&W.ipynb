```python
import random

# Convert string fields to lists
employees = []
for _, row in df.iterrows():
    employees.append({
        "name": row["Name"],
        "available_days": [d.strip() for d in row["Available Days"].split(",")],
        "pref": row["Shift Preference"].lower(),
        "type": row["Fulltime/Part-time"].lower(),
        "assigned": 0  # to track how many shifts assigned
    })

# Create 63 shifts: 7 days × 2 shifts/day × 4.5 slots
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
shift_times = {"morning": "6-2", "evening": "2-10", "night": "10-6"}
shifts = []
id = 1
for day in days:
    for pref, time in shift_times.items():
        for _ in range(3):  # 3 shifts per type per day (7×3×3 = 63)
            shifts.append({"id": f"S{id}", "day": day, "time": time, "type": pref})
            id += 1

# Assign shifts
schedule = []
for shift in shifts:
    candidates = [
        emp for emp in employees
        if shift["day"] in emp["available_days"]
        and (emp["pref"] == "any" or emp["pref"] == shift["type"])
        and ((emp["type"] == "fulltime" and emp["assigned"] < 5) or (emp["type"] == "part-time" and emp["assigned"] < 3))
    ]
    if candidates:
        selected = random.choice(candidates)
        selected["assigned"] += 1
        schedule.append({
            "Shift ID": shift["id"],
            "Day": shift["day"],
            "Time": shift["time"],
            "Shift Type": shift["type"],
            "Assigned To": selected["name"]
        })
    else:
        schedule.append({
            "Shift ID": shift["id"],
            "Day": shift["day"],
            "Time": shift["time"],
            "Shift Type": shift["type"],
            "Assigned To": "Unassigned"
        })

# Export schedule to CSV
schedule_df = pd.DataFrame(schedule)
schedule_df.to_csv("simple_schedule_output.csv", index=False)
print("Schedule saved to simple_schedule_output.csv")
```

```python
# simple_schedule_excel.py
import pandas as pd
import random
from datetime import datetime
# Load employee data from Excel
employees_raw = pd.read_excel("Schedule_aw.xlsx")
# Clean and structure data
employees = []
for _, row in employees_raw.iterrows():
    employees.append({
        "name": row["Name"],
        "available_days": [d.strip() for d in row["Available Days"].split(",")],
        "pref": row["Shift Preference"].lower(),
        "type": row["Fulltime/Part-time"].lower(),
        "assigned": 0,
        "total_hours": 0
    })

# Define all possible shifts with durations (in hours)
shift_definitions = [
    ("6 AM - 2 PM", "morning", ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"], 8),
    ("7 AM - 3 PM", "morning", ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"], 8),
    ("8 AM - 4 PM", "day", ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"], 8),
    ("9 AM - 1 PM", "day", ["Tue", "Wed", "Thu", "Fri"], 4),
    ("2 PM - 10 PM", "evening", ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"], 8),
    ("3 PM - 11 PM", "evening", ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"], 8),
    ("4 PM - 12 PM", "evening", ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"], 8),
    ("7 PM - 2 AM", "night", ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"], 7),
    ("10 PM - 6 AM", "night", ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"], 8),
    ("11 PM - 7 AM", "night", ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"], 8),
    ("11 AM - 7 PM", "day", ["Sun"], 8)
]

# Generate 63 shift slots from available definitions
shifts = []
shift_id = 1
for time, category, days_list, duration in shift_definitions:
    for day in days_list:
        if len(shifts) < 63:
            shifts.append({"id": f"S{shift_id}", "day": day, "time": time, "type": category, "duration": duration})
            shift_id += 1

# Assign shifts
schedule = []
for shift in shifts:
    # Find eligible candidates
    candidates = [
        emp for emp in employees
        if shift["day"] in emp["available_days"]
        and (emp["pref"] == "any" or emp["pref"] == shift["type"])
        and ((emp["type"] == "fulltime" and emp["assigned"] < 4) or (emp["type"] == "part-time" and emp["assigned"] < 3))
    ]
    if candidates:
        selected = random.choice(candidates)
        selected["assigned"] += 1
        selected["total_hours"] += shift["duration"]
        schedule.append({
            "Shift ID": shift["id"],
            "Day": shift["day"],
            "Time": shift["time"],
            "Shift Type": shift["type"],
            "Hours": shift["duration"],
            "Assigned To": selected["name"]
        })
    else:
        schedule.append({
            "Shift ID": shift["id"],
            "Day": shift["day"],
            "Time": shift["time"],
            "Shift Type": shift["type"],
            "Hours": shift["duration"],
            "Assigned To": "Unassigned"
        })

# Export the schedule
schedule_df = pd.DataFrame(schedule)
schedule_df.to_csv("simple_schedule_output_V1.csv", index=False)
print("Schedule saved to simple_schedule_output.csv")

# Summarize hours per person and save
summary = pd.DataFrame([{ "Employee": emp["name"], "Total Hours": emp["total_hours"] } for emp in employees])
summary.to_csv("employee_total_hours.csv", index=False)
print("Total hours per employee saved to employee_total_hours.csv")
```
